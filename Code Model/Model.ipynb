```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
```

```python
# Initialize the model
model = Sequential()
```

```python
# Add Convolutional layers
model.add(Conv2D(32, (7, 7), input_shape=(256, 256, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (7, 7), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (7, 7), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

```

```python
# Flatten the layers
model.add(Flatten())
```

```python
# Add Fully Connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1.0/255)

```

```python
# Load the training and test data
train_data = train_datagen.flow_from_directory('drive/MyDrive/Project_DAP391m_group2/Data/Data_detected/Train_Detected',
                                              target_size=(256, 256),
                                              batch_size=32,
                                              class_mode='binary')

test_data = test_datagen.flow_from_directory('drive/MyDrive/Project_DAP391m_group2/Data/Data_detected/Test_Detected',
                                            target_size=(256, 256),
                                            batch_size=32,
                                            class_mode='binary')

```

```python
# Train the model
model.fit(train_data, epochs=500, validation_data=test_data)
```

```python
# Save the trained model
model.save('drive/MyDrive/Project_DAP391m_group2/Data/Data_detected/cheating_detection_model.h5')
```
